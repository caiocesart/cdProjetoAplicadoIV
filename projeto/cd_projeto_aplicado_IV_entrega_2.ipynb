# **Referencial Teórico**

A previsão de carga elétrica é um problema clássico de séries temporais, amplamente estudado na literatura devido à sua relevância para o planejamento energético, operação de sistemas elétricos e definição de políticas de consumo eficiente. Esse problema pode ser abordado por diferentes classes de modelos, que variam desde métodos estatísticos tradicionais até abordagens baseadas em aprendizado de máquina e deep learning.

Entre os métodos clássicos, destacam-se os modelos autorregressivos integrados de médias móveis (ARIMA) e suas variações sazonais (SARIMA). Esses modelos são amplamente utilizados por sua interpretabilidade e capacidade de capturar estruturas lineares de dependência temporal (BOX et al., 2015). No entanto, apresentam limitações na modelagem de padrões não lineares e podem ter desempenho reduzido em séries com alta variabilidade ou múltiplas sazonalidades.

Como alternativa, métodos baseados em aprendizado de máquina, como Random Forest e Gradient Boosting, têm sido aplicados à previsão de carga elétrica, explorando variáveis derivadas da série temporal, como defasagens (lags) e indicadores sazonais (HONG; FAN, 2016). Esses modelos conseguem capturar relações não lineares e interações complexas, porém dependem de uma etapa cuidadosa de engenharia de atributos e não modelam explicitamente a dependência temporal.

Mais recentemente, modelos de deep learning, como redes neurais recorrentes (RNN) e especialmente Long Short-Term Memory (LSTM), têm sido utilizados para previsão de carga elétrica. Esses modelos são capazes de capturar dependências de longo prazo e padrões complexos diretamente da série temporal (HOCHREITER; SCHMIDHUBER, 1997). Apesar do potencial de desempenho superior, apresentam maior custo computacional, menor interpretabilidade e exigem maior volume de dados e ajuste de hiperparâmetros.

Além disso, abordagens híbridas têm sido propostas na literatura, combinando modelos estatísticos e de aprendizado de máquina, buscando aproveitar as vantagens de cada técnica (ZHANG, 2003). Essas abordagens podem melhorar a acurácia, mas aumentam a complexidade da implementação.

No contexto deste projeto, os principais conceitos envolvidos incluem séries temporais, estacionariedade, tendência, sazonalidade, autocorrelação e validação temporal. A modelagem será realizada a partir de uma série diária agregada, permitindo reduzir ruído e complexidade, ao mesmo tempo em que preserva padrões relevantes para previsão.
